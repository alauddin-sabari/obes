```python
import pandas as pd 
import numpy as np

#Import libraries
from matplotlib_venn import venn2, venn2_circles
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
#file upload

d  = pd.read_csv('../GSE81965_obes_ngt.csv')
d  = d[(abs(d['Log_FoldChange'])>1) & (d['FDR_adjusted_P.value'] <0.05)]

d1 = pd.read_csv('colorectal_cancer/GSE66216_colorectal_cancer.csv')
d1 = d1[(abs(d1['Log_FoldChange'])>1) & (d1['FDR_adjusted_P.value'] <0.05)]

d3 = pd.read_csv('Thyroid_cancer/GSE114069_Thyroid_cancercsv.csv')
d3 = d3[(abs(d3['Log_FoldChange'])>1) & (d3['FDR_adjusted_P.value'] <0.05)]

d4 = pd.read_csv('Hypertension/GSE144274_idiopathic_pulmonary_hypertension.csv')
d4 = d4[(abs(d4['Log_FoldChange'])>1) & (d4['FDR_adjusted_P.value'] <0.05)]

# d5 = pd.read_csv('Osteoarthritis/GSE114007_osteoarthritis.csv')
# d5 = d5[(abs(d4['Log_FoldChange'])>1) & (d5['FDR_adjusted_P.value'] <0.05)]

# d6 = pd.read_csv('Coronary_heart_disease/GSE85565_Coronary_heart_disease.csv')
# d6 = d6[(abs(d6['Log_FoldChange'])>1) & (d6['FDR_adjusted_P.value'] <0.05)]

# d7= pd.read_csv('Gallbladder_cancer/GSE139682_Gallbladder_cancer.csv')
# d7 = d7[(abs(d7['Log_FoldChange'])>1) & (d7['FDR_adjusted_P.value'] <0.05)]
```

Dataset1<br>https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE81965

```python
up = d[d['Log_FoldChange']>1]
up = up.drop_duplicates(subset='Gene_symbol')

set1up = set()
for i in up['Gene_symbol'].values:
    set1up.add(i)
    
#down
down = d[d['Log_FoldChange']<-1]
down = down.drop_duplicates(subset='Gene_symbol')
set1down = set()
for i in down['Gene_symbol'].values:
    set1down.add(i)
```

Dataset2<br>https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE66216

```python
up = d1[d1['Log_FoldChange']>1]
up = up.drop_duplicates(subset='Gene_symbol')

set2up = set()
for i in up['Gene_symbol'].values:
    set2up.add(i)
    
#down
down = d1[d1['Log_FoldChange']<-1]
down = down.drop_duplicates(subset='Gene_symbol')
set2down = set()
for i in down['Gene_symbol'].values:
    set2down.add(i)
```

```python
%matplotlib inline
#venn2(subsets = (30, 10, 5), set_labels = ('Group A', 'Group B'))
venn2([set1up,set2up])
plt.title('obesity vs colorectal cancer (upregulated gene)')
```

```python
%matplotlib inline
v=venn2(subsets=(set1down,set2down),set_labels=('obesity up','colorectal up'), set_colors=('purple', 'skyblue'), alpha = 0.7)
#venn2_circles()
plt.title('obesity vs colorectal cancer (downregulated gene)')
```

```python
import matplotlib_venn as vplt
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.subplot(1, 2, 1)
plt.title('Subplot 1')
v = vplt.venn2(subsets=(10,10,1), set_labels=('A', 'B'), set_colors=('w', 'w'))
c = vplt.venn2_circles(subsets=(2, 2, 1), linestyle='solid')
c[0].set_radius(0.32)
c[1].set_radius(0.32)
c[0].set_lw(2.0)
c[1].set_lw(2.0)
c[0].set_color('green')
c[1].set_color('red')
c[0].set_alpha(0.5)
c[1].set_alpha(0.5)
c[0].set_edgecolor('black')
c[1].set_edgecolor('black')

plt.subplot(1, 2, 2)
plt.title('Subplot 2')
v = vplt.venn2(subsets=(10,1000,1), set_labels=('A', 'B'), set_colors=('w', 'w'))
v.get_label_by_id('10').set_x(0.25)
v.get_label_by_id('01').set_x(-.25)
v.get_label_by_id('11').set_x(0)
v.set_labels[0].set_position((-0.22, -0.45))
v.set_labels[1].set_position((0.25, -0.45))


c = vplt.venn2_circles(subsets=(2, 2, 1), linestyle='solid')
c[0].set_radius(0.32)
c[1].set_radius(0.32)
c[0].set_lw(2.0)
c[1].set_lw(2.0)
c[0].set_color('green')
c[1].set_color('red')
c[0].set_alpha(0.5)
c[1].set_alpha(0.5)
c[0].set_edgecolor('black')
c[1].set_edgecolor('black')
```
