```python
import pandas as pd 
import numpy as np

#Import libraries
from matplotlib_venn import venn2, venn2_circles
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

Dataset's link:<br>https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE58933

```python
up = pd.read_csv('Obesity/appended_UP_Down_obes/up_appended_obes__.csv')
#up
down = pd.read_csv('Obesity/appended_UP_Down_obes/down_appended_obes__.csv')
```

```python
up_down_obes = pd.concat([up,down])
up_down_obes.shape
```

```python
up_down_obes = up_down_obes.drop_duplicates(subset='Gene_symbol')
up_down_obes.shape
```

```python

up = up.drop_duplicates(subset='Gene_symbol')


set1up = set()

for i in up['Gene_symbol'].values:
 
    set1up.add(i)
 
 

 
down = down.drop_duplicates(subset='Gene_symbol')
 
set1down = set()
 
for i in down['Gene_symbol'].values:

    set1down.add(i)
```

```python
up = pd.read_csv('Bipolar_disorder/appended_UP_Down_/up_appended_BD.csv')
down = pd.read_csv('Bipolar_disorder/appended_UP_Down_/down_appended_BD.csv')
```

```python

up = up.drop_duplicates(subset='Gene_symbol')


set2up = set()

for i in up['Gene_symbol'].values:
 
    set2up.add(i)
 

 
down = down.drop_duplicates(subset='Gene_symbol')
 
set2down = set()
 
for i in down['Gene_symbol'].values:

    set2down.add(i)
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_up": set1up,
    "Bipolar_up": set2up
    }
venn(musicians)                 #25->20  421 - >557
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_down": set1down,
    "Bipolar_down": set2down
    }
venn(musicians)                     # 4 ->4
```
