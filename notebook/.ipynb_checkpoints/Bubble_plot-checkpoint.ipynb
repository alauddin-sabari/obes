```python
import pandas as pd 
import numpy as np

#Import libraries
from matplotlib_venn import venn2, venn2_circles
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

Dataset's link:<br>https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE58933

```python
d  = pd.read_csv('Obesity/Data/appended_these_datasets.csv')
d  = d[(abs(d['Log_FoldChange'])>1) & (d['PValue'] <0.05)]
up = d[d['Log_FoldChange']>1]

up = up.drop_duplicates(subset='Gene_symbol')


set1up = set()

for i in up['Gene_symbol'].values:
 
    set1up.add(i)
 
 
down = d[d['Log_FoldChange']<-1]
 
down = down.drop_duplicates(subset='Gene_symbol')
 
set1down = set()
 
for i in down['Gene_symbol'].values:

    set1down.add(i)
```

```python
d1  = pd.read_csv('Bipolar_disorder/Data/appended_BD.csv')
d1  = d1[(abs(d1['Log_FoldChange'])>1) & (d1['PValue'] <0.05)]
#d1  = d1[(abs(d['Log_FoldChange'])>1) & (d1['FDR_adjusted_P.value'] <0.05)]

up = d1[d1['Log_FoldChange']>1]

up = up.drop_duplicates(subset='Gene_symbol')


set2up = set()

for i in up['Gene_symbol'].values:
 
    set2up.add(i)
 
 
down = d1[d1['Log_FoldChange']<-1]
 
down = down.drop_duplicates(subset='Gene_symbol')
 
set2down = set()
 
for i in down['Gene_symbol'].values:

    set2down.add(i)
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_up": set1up,
    "Bipolar_up": set2up
    }
venn(musicians)
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_down": set1down,
    "Bipolar_down": set2down
    }
venn(musicians)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
penguins_data="https://raw.githubusercontent.com/datavizpyr/data/master/palmer_penguin_species.tsv"
```

```python
penguins_df = pd.read_csv(penguins_data, sep="\t")
penguins_df.head()
```

```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.scatterplot(x="culmen_length_mm", 
                y="culmen_depth_mm",
                size="body_mass_g",  #logfold/pvalue
                sizes=(20,500),
                alpha=0.5,
                hue="sex",
                data=penguins_df)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
# Put the legend out of the figure
#plt.legend(bbox_to_anchor=(1.01, 0.54),  borderaxespad=0.)
plt.xlabel("Culmen Length (mm)")
plt.ylabel("Culmen Depth (mm)")
plt.title("Bubble plot with Colors in Seaborn")
plt.tight_layout()
 
plt.savefig("Bubble_plot_Seaborn_color_by_variable_Seaborn_scatterplot.png",
                    format='png',dpi=150)
```
