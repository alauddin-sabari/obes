```python
up = pd.read_csv('Bipolar_disorder/appended_UP_Down_/up_appended_BD.csv')
down = pd.read_csv('Bipolar_disorder/appended_UP_Down_/down_appended_BD.csv')
```

```python

up = up.drop_duplicates(subset='Gene_symbol')


set2up = set()

for i in up['Gene_symbol'].values:
 
    set2up.add(i)
 

 
down = down.drop_duplicates(subset='Gene_symbol')
 
set2down = set()
 
for i in down['Gene_symbol'].values:

    set2down.add(i)
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_up": set1up,
    "Bipolar_up": set2up
    }
venn(musicians)                 #25->20  421 - >557
```
