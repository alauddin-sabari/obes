```python
up = pd.read_csv('Schizophrenia/appended_UP_Down_/up_appended_SZ.csv')
down = pd.read_csv('Schizophrenia/appended_UP_Down_/down_appended_SZ.csv')
```

```python

up = up.drop_duplicates(subset='Gene_symbol')


set23up = set()

for i in up['Gene_symbol'].values:
 
    set23up.add(i)
 
 

 
down = down.drop_duplicates(subset='Gene_symbol')
 
set23down = set()
 
for i in down['Gene_symbol'].values:

    set23down.add(i)
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_up": set1up,
    "Schizophrenia_up": set23up
    }
venn(musicians)                   #15->11
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_down": set1down,
    "Schizophrenia_down": set23down
    }
venn(musicians)      #30 - >18
```
