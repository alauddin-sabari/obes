```python
up = pd.read_csv('Obesity/appended_UP_Down_obes/up_appended_obes__.csv')
#up
down = pd.read_csv('Obesity/appended_UP_Down_obes/down_appended_obes__.csv')
```

```python
up_down_obes = pd.concat([up,down])
up_down_obes.shape
```

```python
up_down_obes = up_down_obes.drop_duplicates(subset='Gene_symbol')
up_down_obes.shape
```

```python

up = up.drop_duplicates(subset='Gene_symbol')


set1up = set()

for i in up['Gene_symbol'].values:
 
    set1up.add(i)
 
 

 
down = down.drop_duplicates(subset='Gene_symbol')
 
set1down = set()
 
for i in down['Gene_symbol'].values:

    set1down.add(i)
```

# Major_depression

```python
up = pd.read_csv('Major_depression/appended_UP_Down_/up_appended_MD.csv')

down = pd.read_csv('Major_depression/appended_UP_Down_/down_appended_MD.csv')
```

```python

set21up = set()

for i in up['Gene_symbol'].values:
 
    set21up.add(i)
 
 

down = down.drop_duplicates(subset='Gene_symbol')
 
set21down = set()
 
for i in down['Gene_symbol'].values:

    set21down.add(i)
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_up": set1up,
    "Major_depression_up": set21up
    }
venn(musicians)                    #4->3
```

```python
from venn import venn
%matplotlib inline

musicians = {
    "Obesity_down": set1down,
    "Major_depression": set21down
    }
venn(musicians)        #8 > 
```
